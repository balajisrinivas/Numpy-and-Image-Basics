## Images

Images are usually 3 channels of Red, Green and Blue. In short, we can call them RGB.
Each channel takes value from 0 to 255, where 0 being dark and 255 being bright.

Say, if (R, G, B) = (0, 0, 0) in a picture, the resulting color is black.
Similarly, if (R, G, B) = (255, 255, 255), the resulting color is white.

Essentially, any color is a combination of these three colors -

![alt text](https://github.com/balajisrinivas/Numpy-and-Image-Basics/blob/master/rgb.PNG "RGB picture")

Further, diving deep, any image is a three dimensional array with with pixels of height, width and the color channels.

![alt text](https://github.com/balajisrinivas/Numpy-and-Image-Basics/blob/master/color.PNG "Parrot picture")
